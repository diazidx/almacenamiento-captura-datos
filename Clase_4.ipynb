### Ejercicio 2

Extraer notebooks de pc factory

```python
r = requests.get("https://www.pcfactory.cl/notebooks?categoria=735&papa=636")
res = BeautifulSoup(r.text,"html.parser")
print(res.title.text)
```

```python
npaginas = res.find('span', {'id': 'pag_ter2'}).text
print("Número de páginas:", npaginas)
```

```python
bd = []

for n in range(int(npaginas)):
    req = requests.get("https://www.pcfactory.cl/notebooks?categoria=735&papa=636&pagina=" + str(n + 1))

    res = BeautifulSoup(req.text,"html.parser")

    notebooks = res.find_all('div', {'class': 'wrap-caluga-matrix'})

    for n in notebooks:
        marca = n.find('span', {'class': 'marca'}).text
        nombre = n.find('span', {'class': 'nombre'}).text
        precio = n.find('meta', {'itemprop': 'price'})['content']

        bd.append({"brand": marca, "name": nombre, "price": int(float(precio))})
```

```python
import pandas as pd

df = pd.DataFrame(bd)
df
```

### Ejercicio 3

Extraer productos y precios de linio

```python
import requests
productos = []
for i in range(3):
    r = requests.get("https://www.linio.cl/c/computacion/pc-portatil?page=" + str(i+1))
    res = BeautifulSoup(r.text,"html5lib")
    products = res.find_all('div', {'class':'catalogue-product row'})
    for p in products:
        title = p.find('meta', {'itemprop':'name'})
        price = p.find('meta', {'itemprop':'price'})
        if title:
            print(title['content'])
        if price:
            print(price['content'])
        productos.append({'title':title['content'], 'price':price['content']})
```

```python
import pandas as pd
pd.DataFrame(productos)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.implicitly_wait(3)

driver.get('https://google.cl')

elem = driver.find_element_by_name('q')
elem.clear()
elem.send_keys("Magister en data science")
elem.send_keys(Keys.RETURN)

elems = driver.find_elements_by_css_selector('div.g')
for e in elems:
    try:
        print(e.find_element_by_tag_name('h3').text)
        print(e.find_element_by_tag_name('cite').text)
        print()
    except:
        print("No encontré elementos")
        print()
    
driver.close()
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.implicitly_wait(3)

driver.get('https://es.wikipedia.org') 

elem = driver.find_element_by_name('search')
elem.clear()

#b. Ingresar en el input de búsqueda el string “paises por superficie” y presionar enter.
elem.send_keys('paises por superficie')
elem.send_keys(Keys.RETURN)

columna    = [] 
territorio = [] 
superficie = [] 

h = driver.find_elements_by_xpath("//tr/th")
for i in h:
    columna.append(i.text)

    
elemento_1 = driver.find_elements_by_xpath("//tr/td[2]")
for e in elemento_1:
    territorio.append(e.text)
    
elemento_2 = driver.find_elements_by_xpath("//tr/td[3]")

#d. Extraer la lista de países y territorios con su superficie desde la página de resultados
#(no incluir Planeta Tierra, Océanos y Tierras emergidas). (2puntos)

for a in elemento_2:
    superficie.append(a.text)

#Cierro el navegador
driver.close()
```
