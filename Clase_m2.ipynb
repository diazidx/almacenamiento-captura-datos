```python
#calcular cuanto se demora un codigo en ejecutar
import timeit
```

```python
code1 = """
import pandas as pd
import sqlite3
conn = sqlite3.connect('data/chinook.db')
cur = conn.cursor()
cur.execute('SELECT * FROM albums a JOIN artists ar ON a.artistId = ar.artistId')
albums = cur.fetchall()
albums
pd.DataFrame(albums, columns=['AlbumID', 'Title', 'ArtistId', 'ArtistId', 'Name'])
"""
time1 = timeit.timeit(code1, number = 10000)
print(time1)
```

```python
code2 = """
import pandas as pd
import sqlite3
conn = sqlite3.connect('data/chinook.db')
cur = conn.cursor()
pd.read_sql('SELECT * FROM albums a JOIN artists ar ON a.artistId = ar.artistId', conn)
"""
time2 = timeit.timeit(code2, number = 10000)
print(time2)
```

```python
df1 = pd.read_sql('SELECT * FROM albums', conn)
df2 = pd.read_sql('SELECT * FROM artists', conn)
df3 = df1.merge(df2, how = 'inner', on = 'ArtistId')
df3
```

```python
code3 = """
import pandas as pd
import sqlite3
conn = sqlite3.connect('data/chinook.db')
df1 = pd.read_sql('SELECT * FROM albums', conn)
df2 = pd.read_sql('SELECT * FROM artists', conn)
df1.merge(df2, how = 'inner', on = 'ArtistId')
"""
time3 = timeit.timeit(code3, number = 10000)
print(time3)
```
